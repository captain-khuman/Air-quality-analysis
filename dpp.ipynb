```python
import numpy as np
lower_interval = int(np.floor( df3_scaled['overall'].min()))
upper_interval = int(np.ceil( df3_scaled['overall'].max()))
range=1
interval_length = range / 5
print(lower_interval)
print(upper_interval)
print(interval_length)
```

```python
total_bins = np.arange(lower_interval, upper_interval+interval_length, interval_length)
print(total_bins)
```
